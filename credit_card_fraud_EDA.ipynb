```python
#load data

import pandas as pd
df = pd.read_csv("credit_card_data.csv")
df.head()
```

```python
#check basic info

df.shape   #rows and column
```

```python
df.info()  
```

```python
df.describe()
```

```python
#convert timestamp to datetime

df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
#Add hour of transaction

df['hour'] = df['timestamp'].dt.hour
```

```python
#Create fraud amount column

df['fraud_amount'] = df['amount'] * df['is_fraud']
```

```python
#create high value flag

df['high_value'] = df['amount'] > 50000
```

```python
df['is_fraud'].value_counts(normalize=True) * 100
```

```python
#save cleaned file

df.to_csv("cleaned_credit_card_data.csv", index=False)
```
